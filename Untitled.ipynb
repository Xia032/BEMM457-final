```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("LondonBikeJourneyAug2023.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.drop_duplicates()
```

```python
df.isna().sum()
```

```python
df = df.dropna()
```

```python
df["Start date"] = pd.to_datetime(df["Start date"])
```

```python
df["date"] = df["Start date"].dt.date
df["hour"] = df["Start date"].dt.hour
```

```python
df = df[["Start date", "date", "hour", "Start station number", "End station number"]]
```

```python
daily_counts = df.groupby("date").size()

plt.figure(figsize=(10,5))
plt.plot(daily_counts.index, daily_counts.values)
plt.title("Daily Rentals")
plt.xlabel("Date")
plt.ylabel("Number of Rentals")
plt.grid(True)
plt.show()
```

```python
hourly_counts = df.groupby("hour").size()

plt.figure(figsize=(10,5))
plt.plot(hourly_counts.index, hourly_counts.values)
plt.title("Hourly Rental")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Rentals")
plt.grid(True)
plt.show()
```

```python
import statsmodels.api as sm
```

```python
hourly = df.groupby(["date", "hour"]).size().reset_index(name="rentals")
```

```python
X = sm.add_constant(hourly[["hour"]])
y = hourly["rentals"]    
model = sm.OLS(y, X).fit()
```

```python
print("R-squared (simple linear regression):", model.rsquared)
print(model.params)
```

```python
actual_mean = hourly.groupby("hour")["rentals"].mean()
hourly["predicted"] = model.predict(X)
pred_mean = hourly.groupby("hour")["predicted"].mean()
```

```python
plt.figure(figsize=(8, 4))
plt.plot(actual_mean.index, actual_mean.values, label="Actual average rentals")
plt.plot(pred_mean.index, pred_mean.values, label="Predicted average rentals")

plt.xlabel("Hour of day")
plt.ylabel("Average rentals")
plt.title("Actual vs Predicted Average Hourly Bike Rentals")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
avg_hourly = hourly.groupby("hour")["rentals"].mean().reset_index()

# train model on avg_hourly
X = sm.add_constant(avg_hourly[["hour"]])
y = avg_hourly["rentals"]
model = sm.OLS(y, X).fit()

# generate predictions
avg_hourly["PredictedAvg"] = model.predict(sm.add_constant(avg_hourly[["hour"]]))

avg_hourly.to_csv("hourly_comparison.csv", index=False)
```
